```python
data['date'] = pd.to_datetime(data['date'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Gráfico de distribuição por raça
plt.figure(figsize=(10,6))
sns.countplot(data=data, x='race', order=data['race'].value_counts().index,palette='deep')
plt.title('Distribution of shooting per race')
plt.xlabel('Race')
plt.ylabel('Count')
plt.show()
```

The data base goes only half of 2020, thats why is so down. But we can see that the shooting are decreasing over the years.

```python
# Gráfico de tendências ao longo do tempo
data['year'] = data['date'].dt.year
plt.figure(figsize=(10,6))
sns.countplot(data=data, x='year',palette='deep')
plt.title('Tendency of shooting over the years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
not_fleeing = data[data['flee'] == 'Not fleeing'].copy()

fig, ax = plt.subplots(2,1, figsize=(8,8))

sns.countplot(ax=ax[0], data=not_fleeing, y='threat_level',palette='deep')
sns.countplot(ax=ax[1], data=not_fleeing, y='armed',
              order=not_fleeing.armed.value_counts().iloc[:5].index,palette='deep')

plt.show()
```
